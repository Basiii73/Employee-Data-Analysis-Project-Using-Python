```python
import pandas as pd


url = 'https://docs.google.com/spreadsheets/d/1VP9BE_eI2yl6uUHSm4mGiiwjRdoqCqnkcIjsv5Q2ex4/export?format=csv'
df = pd.read_csv(url)
```

```python
#Analysis Tasks
```

```python
import numpy as np


np.random.seed(42) 
df['height'] = np.random.randint(150, 181, size=len(df))
```

```python
# Count employees per team
team_distribution = df['Team'].value_counts()

# Calculate percentage
team_percentage = (team_distribution / len(df)) * 100
```

```python
# Define age bins and labels
bins = [20, 30, 40, 50, 60, 70]
labels = ['21-30', '31-40', '41-50', '51-60', '61-70']

# Create age groups
df['age_group'] = pd.cut(df['Age'], bins=bins, labels=labels, right=False)

# Count employees per age group
age_group_distribution = df['age_group'].value_counts().sort_index()
```

```python
# Total salary per team
team_salary = df.groupby('Team')['Salary'].sum()

# Total salary per position
position_salary = df.groupby('Position')['Salary'].sum()
```

```python
# Calculate correlation
age_salary_correlation = df['Age'].corr(df['Salary'])
```

```python
#Graphical Representation:
```

```python
import matplotlib.pyplot as plt

# Pie chart for team distribution
team_percentage.plot.pie(autopct='%1.1f%%', figsize=(8, 8), title='Employee Distribution by Team')
plt.ylabel('') 
plt.show()
```

```python
# Count employees per position
position_distribution = df['Position'].value_counts()

# Plot the bar chart
position_distribution.plot.bar(figsize=(10, 6), title='Employee Distribution by Position')
plt.xlabel('Position')
plt.ylabel('Number of Employees')
plt.show()

```

```python
# Bar chart for age group distribution
age_group_distribution.plot.bar(figsize=(10, 6), title='Employee Distribution by Age Group')
plt.xlabel('Age Group')
plt.ylabel('Number of Employees')
plt.show()
```

```python
# Bar chart for team salary expenditure
team_salary.plot.bar(figsize=(10, 6), title='Total Salary Expenditure by Team')
plt.xlabel('Team')
plt.ylabel('Total Salary')
plt.show()

# Bar chart for position salary expenditure
position_salary.plot.bar(figsize=(10, 6), title='Total Salary Expenditure by Position')
plt.xlabel('Position')
plt.ylabel('Total Salary')
plt.show()
```

```python
# Scatter plot for age vs. salary
plt.figure(figsize=(10, 6))
plt.scatter(df['Age'], df['Salary'], alpha=0.5)
plt.title('Age vs. Salary')
plt.xlabel('Age')
plt.ylabel('Salary')
plt.show()
```
